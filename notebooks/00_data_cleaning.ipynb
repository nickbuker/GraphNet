# Data Cleaning

Removes duplicate entries from the Bradley Double Plus Good Melting Point Dataset and saves the cleaned data to `data/processed/deduplicated.parquet`.

```python
import matplotlib.pyplot as plt
import polars as pl

plt.style.use("ggplot")
```

```python
df = pl.read_csv("../data/4/BradleyDoublePlusGoodMeltingPointDataset.csv")
df.head(10)
```

```python
df.height
```

```python
df["name"].n_unique()
```

```python
df["smiles"].n_unique()
```

```python
df.filter(df["name"].is_duplicated())
```

```python
df = df.remove(pl.col("csid") == 389934)
df.height
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df["mpC"], bins=15)
plt.title("Distribution of errors")
plt.tight_layout()
plt.show();
```

```python
df.select(["key", "name", "smiles", "mpC"]).write_parquet("../data/processed/deduplicated.parquet")
```
